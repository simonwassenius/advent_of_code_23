```python
# Read the contents of the file and split it into lines
with open("AOC2023_14.txt", "r") as file:
    f = [list(line) for line in file.read().split('\n')]

# Iterate over the rows in reverse order
looping = True
while looping:
    looping = False
    for y in range(len(f)-1):
        for ix, char in enumerate(f[-1-y]):
            # Check if the current cell is 'O' and the cell above is '.'
            if char == "O" and f[-2-y][ix] == ".":
                # Move the 'O' down one row
                f[-1-y][ix] = "."
                f[-2-y][ix] = "O"
                looping = True
# count the score
i = 1
ans = 0
for ixx, row in enumerate(reversed(f)):
    for char in row:
        if char == "O":
            ans += i
            # print("Char = {}, Row =  {}, i = {}, ans = {}".format(char, ixx, i, ans))
    i += 1
    print("".join(row))

ans
```

```python
def move_ones(grid, direction):
    shifts = {"north": (-1, 0), "west": (0, -1), "south": (1, 0), "east": (0, 1)}

    for y in range(len(grid)-1, -1, -1):
        for x in range(len(grid[0])-1, -1, -1):
            if grid[y][x] == "O":
                move_max(grid, y, x, shifts[direction])

def move_max(grid, y, x, shift):
    new_y, new_x = y + shift[0], x + shift[1]

    while 0 <= new_y < len(grid) and 0 <= new_x < len(grid[0]) and (grid[new_y][new_x] == "." or grid[new_y][new_x] == "O"):
        grid[y][x], grid[new_y][new_x] = grid[new_y][new_x], grid[y][x]
        y, x = new_y, new_x
        new_y, new_x = y + shift[0], x + shift[1]

# Read the contents of the file and split it into lines
with open("AOC2023_14.txt", "r") as file:
    f = [list(line) for line in file.read().split('\n')]

# Print the initial grid
print("Cycle 0 - Before Movement:")
for row in f:
    print("".join(row))

# Perform the first 3 cycles
for cycle in range(1, 4):
    for direction in ["north", "west", "south", "east"]:
        print(f"\nCycle {cycle} - Before Moving 'O's {direction.capitalize()}:")
        for row in f:
            print("".join(row))
        move_ones(f, direction)

# Print the final grid
print("\nFinal Grid:")
for row in f:
    print("".join(row))
```
