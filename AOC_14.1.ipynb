```python
# Read the contents of the file and split it into lines
with open("AOC2023_13.txt", "r") as file:
    f = file.read().split('\n\n')

# f ="""#.##..##.
# ..#.##.#.
# ##......#
# ##......#
# ..#.##.#.
# ..##..##.
# #.#.##.#.

# #...##..#
# #....#..#
# ..##..###
# #####.##.
# #####.##.
# ..##..###
# #....#..#"""
# f = f.split('\n\n')


f = [row.split("\n") for row in f]


ans = 0
for matrix in f:
    count = 0
    multiplier = 100
    print(matrix)
    searching = True
    i = 0
    while searching:
        if matrix[i] == matrix[i+1]:
            print("reflection at {} & {}".format(i, i+1))
            testing = True
            ii = 1
            if (i == 0) or (i == len(matrix)-2):
                testing = False
                count = i+1
            while testing:
                print("index is {}".format(ii))
                if matrix[i-ii] != matrix[i+1+ii]:
                    print("row {} and {}, not matching".format((i-ii), (i+1+ii)))
                    testing = False
                elif (i-ii <= 0) or (i+1+ii >= len(matrix)-1):
                    print("End")
                    count = i+1
                    testing = False
                    i = len(matrix)-2
                ii += 1
            
        if i == len(matrix)-2:
            if count == 0:
                print("Count == 0, Transposing")
                # Transpose and reset
                f_new = []
                for x in range(len(matrix[0])):
                    addition = ""
                    for y in range(len(matrix)):
                        addition += matrix[y][x]
                    f_new.append(addition)

                print(f_new)
                matrix = f_new
                i = -1
                multiplier = 1
            else:
                searching = False
                print("score for this matrix is {}" .format(count*multiplier))
                ans += (count * multiplier)
                print("ans = {}" .format(ans))
        i += 1
    print()
```
