```python
import numpy as np

# Read the contents of the file and split it into lines
with open("AOC2023_10.txt", "r") as file:
    f = file.read().split('\n')

# Find start position
for y, row in enumerate(f):
    for x, char in enumerate(row):
        if char == "S":
            pos = [y, x]

# Initialize loop variables
loop = True
i = 1
pos_path = [pos]

# Start movement based on the initial character
if f[pos[0] - 1][pos[1]] in {"|", "F"}:
    pos[0] -= 1
    prev = "D"
elif f[pos[0]][pos[1] + 1] in {"-", "J"}:
    pos[1] += 1
    prev = "L"
elif f[pos[0] + 1][pos[1]] in {"|", "J"}:
    pos[0] += 1
    prev = "U"
elif f[pos[0]][pos[1] - 1] in {"-", "F"}:
    pos[1] -= 1
    prev = "R"
pos_path.append([pos[0], pos[1]])

while loop:
    # Movement conditions
    if f[pos[0]][pos[1]] == "|" and prev == "D":
        pos[0] -= 1
        prev = "D"
        pos_path.append([pos[0], pos[1]])
    elif f[pos[0]][pos[1]] == "|" and prev == "U":
        pos[0] += 1
        prev = "U"
        pos_path.append([pos[0], pos[1]])
    elif f[pos[0]][pos[1]] == "-" and prev == "L":
        pos[1] += 1
        prev = "L"
        pos_path.append([pos[0], pos[1]])
    elif f[pos[0]][pos[1]] == "-" and prev == "R":
        pos[1] -= 1
        prev = "R"
        pos_path.append([pos[0], pos[1]])
    elif f[pos[0]][pos[1]] == "F" and prev == "D":
        pos[1] += 1
        prev = "L"
        pos_path.append([pos[0], pos[1]])
    elif f[pos[0]][pos[1]] == "F" and prev == "R":
        pos[0] += 1
        prev = "U"
        pos_path.append([pos[0], pos[1]])
    elif f[pos[0]][pos[1]] == "J" and prev == "U":
        pos[1] -= 1
        prev = "R"
        pos_path.append([pos[0], pos[1]])
    elif f[pos[0]][pos[1]] == "J" and prev == "L":
        pos[0] -= 1
        prev = "D"
        pos_path.append([pos[0], pos[1]])
    elif f[pos[0]][pos[1]] == "L" and prev == "R":
        pos[0] -= 1
        prev = "D"
        pos_path.append([pos[0], pos[1]])
    elif f[pos[0]][pos[1]] == "L" and prev == "U":
        pos[1] += 1
        prev = "L"
        pos_path.append([pos[0], pos[1]])
    elif f[pos[0]][pos[1]] == "7" and prev == "D":
        pos[1] -= 1
        prev = "R"
        pos_path.append([pos[0], pos[1]])
    elif f[pos[0]][pos[1]] == "7" and prev == "L":
        pos[0] += 1
        prev = "U"
        pos_path.append([pos[0], pos[1]])

    # Check for the end condition
    if f[pos[0]][pos[1]] == "S":
        loop = False

    i += 1

ans_1 = int(np.ceil(i / 2))


def shoelace_formula(vertices):
    n = len(vertices)
    area = 0.0

    for i in range(n):
        x1, y1 = vertices[i]
        x2, y2 = vertices[(i + 1) % n]  # To ensure that the last vertex connects to the first one

        area += (x1 * y2) - (x2 * y1)

    area = abs(area) / 2.0
    return area

area = shoelace_formula(pos_path)
inside = area - len(pos_path) / 2 + 1
ans_2 = int(np.ceil(inside))

print(f"Answer 1: {ans_1}")
print(f"Answer 2: {ans_2}")
```
