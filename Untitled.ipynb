```python
import pandas as pd
import numpy as np
import yfinance as yf
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
from ta.utils import dropna
from ta.volatility import BollingerBands
from ta.momentum import RSIIndicator, StochasticOscillator
from ta.trend import MACD, SMAIndicator

# Parameters
train_symbols = ['AAPL', 'MSFT', 'GOOGL']  # Training stocks
test_symbol = 'AMZN'  # Testing stock
start_date = '2020-01-01'
end_date = '2023-01-01'
initial_capital = 100000.0
risk_per_trade = 0.01  # Risk 1% of capital per trade
risk_reward_ratio = 2  # 1:2 risk-reward ratio
volatility_window = 14  # For volatility-based position sizing

# Download historical data
def download_data(symbols, start, end):
    data = {}
    for symbol in symbols:
        data[symbol] = yf.download(symbol, start=start, end=end)
    return data

train_data = download_data(train_symbols, start_date, end_date)
test_data = yf.download(test_symbol, start=start_date, end=end_date)

# Feature Engineering
def create_features(data):
    data = dropna(data)
    data['SMA_10'] = SMAIndicator(data['Close'], window=10).sma_indicator()
    data['SMA_50'] = SMAIndicator(data['Close'], window=50).sma_indicator()
    data['RSI'] = RSIIndicator(data['Close'], window=14).rsi()
    data['Stoch_Osc'] = StochasticOscillator(data['High'], data['Low'], data['Close']).stoch()
    data['MACD'] = MACD(data['Close']).macd()
    bb = BollingerBands(data['Close'])
    data['BB_High'] = bb.bollinger_hband()
    data['BB_Low'] = bb.bollinger_lband()
    data['Target'] = np.where(data['Close'].shift(-1) > data['Close'], 1, 0)
    data = data.dropna()
    return data

for symbol in train_symbols:
    train_data[symbol] = create_features(train_data[symbol])
    print(f"Train data for {symbol}:")
    print(train_data[symbol].head())  # Debug print
    print(train_data[symbol].tail())  # Debug print
test_data = create_features(test_data)
print(f"Test data for {test_symbol}:")
print(test_data.head())  # Debug print
print(test_data.tail())  # Debug print

# Combine training data
combined_train_data = pd.concat([train_data[symbol] for symbol in train_symbols])

# Check for empty target or features
if combined_train_data.empty or combined_train_data['Target'].isnull().all():
    raise ValueError("Combined training data is empty or target column is missing values")

# Model Training with Hyperparameter Tuning
features = combined_train_data.columns.difference(['Target', 'Adj Close'])
X_train = combined_train_data[features]
y_train = combined_train_data['Target']

if X_train.empty or y_train.empty:
    raise ValueError("Training features or target array is empty")

model = RandomForestClassifier(random_state=42)
param_grid = {
    'n_estimators': [50, 100, 200],
    'max_depth': [5, 10, 20],
    'min_samples_split': [2, 5, 10]
}

grid_search = GridSearchCV(model, param_grid, cv=3, scoring='accuracy')
grid_search.fit(X_train, y_train)
best_model = grid_search.best_estimator_

# Model Testing
X_test = test_data[features]
y_test = test_data['Target']
predictions = best_model.predict(X_test)

accuracy = accuracy_score(y_test, predictions)
print(f"Model Accuracy: {accuracy:.2f}")

# Simulate trading with risk management
test_data['Predictions'] = predictions
test_data['Signal'] = test_data['Predictions'].diff()
test_data['Portfolio Value'] = initial_capital
test_data['Holdings'] = 0

def calculate_position_size(volatility, capital, risk_per_trade):
    return (capital * risk_per_trade) / volatility

for i in range(1, len(test_data)):
    if test_data['Signal'].iloc[i] == 1:
        position_size = calculate_position_size(test_data['Close'].rolling(volatility_window).std().iloc[i], test_data['Portfolio Value'].iloc[i-1], risk_per_trade)
        test_data.loc[test_data.index[i], 'Holdings'] = test_data['Holdings'].iloc[i-1] + position_size
        test_data.loc[test_data.index[i], 'Portfolio Value'] = test_data['Portfolio Value'].iloc[i-1] - (position_size * test_data['Close'].iloc[i])
    elif test_data['Signal'].iloc[i] == -1:
        position_size = test_data['Holdings'].iloc[i-1]
        test_data.loc[test_data.index[i], 'Holdings'] = 0
        test_data.loc[test_data.index[i], 'Portfolio Value'] = test_data['Portfolio Value'].iloc[i-1] + (position_size * test_data['Close'].iloc[i])
    else:
        test_data.loc[test_data.index[i], 'Holdings'] = test_data['Holdings'].iloc[i-1]
        test_data.loc[test_data.index[i], 'Portfolio Value'] = test_data['Portfolio Value'].iloc[i-1]

    # Implement stop-loss and take-profit
    current_value = test_data['Portfolio Value'].iloc[i] + (test_data['Holdings'].iloc[i] * test_data['Close'].iloc[i])
    entry_value = test_data['Portfolio Value'].iloc[i-1]
    stop_loss_value = entry_value * stop_loss_threshold
    take_profit_value = entry_value * take_profit_threshold

    if current_value < stop_loss_value:
        test_data.loc[test_data.index[i], 'Holdings'] = 0
        test_data.loc[test_data.index[i], 'Portfolio Value'] = current_value
    elif current_value > take_profit_value:
        test_data.loc[test_data.index[i], 'Holdings'] = 0
        test_data.loc[test_data.index[i], 'Portfolio Value'] = current_value

test_data['Total Value'] = test_data['Portfolio Value'] + (test_data['Holdings'] * test_data['Close'])

# Final portfolio value
final_portfolio_value = test_data['Total Value'].iloc[-1]
print(f"Final Portfolio Value: ${final_portfolio_value:.2f}")

# Plot the portfolio value
plt.figure(figsize=(14, 7))
plt.plot(test_data['Total Value'], label='Total Portfolio Value')
plt.title(f'Portfolio Value Over Time')
plt.legend(loc='best')
plt.show()
```
